```python

#Simple 2 layers Neural Network
# This network has 2 Inputs, 2 Hidden Neurons and 1 Output
#The program first ask about the number of inputs, number of hidden neurons, 
# number of output , Number of epochs, and the learning rate.
#Once these entered then asks for the inputs separated by SINGLE space
# e.g 1 1 and the3n press ENTER
# The neural network will start learning the weights and 
# updates the weight untill it matches the output
# Once matched the program will be print 
# the number of epoch at which learned the correct output
import numpy as np
from random import randint
import random

# Ask the user for parameters i.e inputs, hidden units and outputs
def Parameters():
    print("Simple Neural Network 3-4-1 (3 input, 4 hidden, 1 output) ")
    input_size = int(input("Enter the number of inputs "))
    hidden_unit_size = int(input("Enter the number of hidden units " ))

    ouput_size = int(input("Enter the number of output " ))
    no_of_epochs = int(input("Enter the number of epochs (iteration) "))
    lr = float(input("Enter learning rate "))
    Initialize(input_size,hidden_unit_size)
    X = Input(input_size)
    return X,no_of_epochs,lr
#Ask the user to input the data i.e x1,x2,,,,xn
def Sigmoid(x):
        return 1/(1+np.exp(-x))

def Input(input_size):
  print (f"Please enter {input_size} inputs(x1,x2,...)")
  X = np.array([int(x) for x in input("Values are 0 and 1, separated by space: ").split(" ")])
  return X
      
# Create the actual neural network structure, Inputs, Hidden Units, Output
def Structure ():
    global input_size 
    global hidden_unit_size
    global ouput_size
    global no_of_epochs
    global lr
    global X
    global first_layer_weights
    global hidden_layer_weights 
    global output
    output =1
    X= []
    first_layer_weights =[]
    hidden_layer_weights= []
    lr =0.1
#Produce all the weights randomly between 0 and 1, i.e w1,w2,w3,...
def Initialize(input,hidden):
    #Weight Formula W = (input+1)*hidden
    No_first_layer_weights = 4*input
    No_hidden_layer_weights = hidden
    for i in range(0,No_first_layer_weights):
        first_layer_weights.append( random.uniform(0,1))
    for i in range(0,No_hidden_layer_weights):
        hidden_layer_weights.append( random.uniform(0,1))
    
    print("First Layer Weights\n",first_layer_weights)
    
    print("Hidden Layer Weights\n",hidden_layer_weights)
         
#This function provide one training sample at a time at the input layer
def Introduce():
    pass
#This function calculate output from the hidden units and 
# overall outputs from the network for each input sample training
# using weighted sum formula i.e wsum = w1*x1+w2*x2+...
def Output():
    X, epochs,lr = Parameters()
    for i in range(epochs):
       
    #Forward Propagation
     y1 = X[0]*first_layer_weights[0]+X[1]*first_layer_weights[1]+X[2]*first_layer_weights[2]
     y2 = X[0]*first_layer_weights[3]+X[1]*first_layer_weights[1]+X[2]*first_layer_weights[5]
     y3= X[0]*first_layer_weights[6]+X[1]*first_layer_weights[7]+X[2]*first_layer_weights[8]
     y4 = X[0]*first_layer_weights[9]+X[1]*first_layer_weights[10]+X[2]*first_layer_weights[11]
     y5 = Sigmoid(y1)
     y6 = Sigmoid(y2)
     y7 = Sigmoid(y3)
     y8 = Sigmoid(y4)
     y = y5*hidden_layer_weights[0]+y6*hidden_layer_weights[1]+y7*hidden_layer_weights[2]+y8*hidden_layer_weights[3]
     y= Sigmoid(y)
     #Update the weights of the network
     WeightUpdate(y,y5,y6,y7,y8,lr,X,first_layer_weights,hidden_layer_weights)
     
     #print(f"iteration number {i}")
     if output-y<=0.001:
         print(f"Predicted Value Matched with the Target after {i+1} iterations")
         break
def WeightUpdate(y,y5,y6,y7,y8,lr,X,first_layer_weights,hidden_layer_weights):
     first_layer_weights[0]=first_layer_weights[0]+lr*(output - y)*X[0]
     first_layer_weights[3]=first_layer_weights[3]+lr*(output - y)*X[0]
     first_layer_weights[6]=first_layer_weights[6]+lr*(output - y)*X[0]
     first_layer_weights[9]=first_layer_weights[9]+lr*(output - y)*X[0]

     first_layer_weights[1]=first_layer_weights[1]+lr*(output - y)*X[1]
     first_layer_weights[4]=first_layer_weights[4]+lr*(output - y)*X[1]
     first_layer_weights[7]=first_layer_weights[7]+lr*(output - y)*X[1]
     first_layer_weights[10]=first_layer_weights[10]+lr*(output - y)*X[1]
     

     first_layer_weights[2]=first_layer_weights[2]+lr*(output - y)*X[2]
     first_layer_weights[5]=first_layer_weights[5]+lr*(output - y)*X[2]
     first_layer_weights[8]=first_layer_weights[8]+lr*(output - y)*X[2]
     first_layer_weights[11]=first_layer_weights[11]+lr*(output - y)*X[2]
     
     


     hidden_layer_weights[0] =hidden_layer_weights[0]+lr*(output-y)*y5
     hidden_layer_weights[1] =hidden_layer_weights[1]+lr*(output-y)*y6
     hidden_layer_weights[2] =hidden_layer_weights[2]+lr*(output-y)*y7
     hidden_layer_weights[3] =hidden_layer_weights[3]+lr*(output-y)*y8



#print the weights
def Print():
     print(first_layer_weights)
     print(hidden_layer_weights)
Structure()
Output()
Print()
```
